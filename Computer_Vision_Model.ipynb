Importing the TensorFlow

```python
import tensorflow as tf
print(tf.__version__)
```

```python
#importing the dataset
mnist = tf.keras.datasets.fashion_mnist
```

```python
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(training_images[0])
print(training_labels[0])
print(training_images[0])
```

```python
training_images = training_images / 255.0
test_images = test_images / 255.0
```

Designing the Model

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(training_images, training_labels, epochs=5)
```

```python
#Testing the model
model.evaluate(test_images, test_labels)
```

# Introduction to Convolutions

```python
#Importing the py libraries
import cv2
import numpy as np
from scipy import datasets
i = datasets.ascent()
```

```python
import matplotlib.pyplot as plt
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(i)
plt.show()
```

```python
i_transformed = np.copy(i)
size_x = i_transformed.shape[0]
size_y = i_transformed.shape[1]
```

Create the convolution matrix

```python
# This filter detects edges nicely
# It creates a filter that only passes through sharp edges and straight lines.
# Experiment with different values for fun effects.
#filter = [ [0, 1, 0], [1, -4, 1], [0, 1, 0]]
# A couple more filters to try for fun!
filter = [ [-1, -2, -1], [0, 0, 0], [1, 2, 1]]
#filter = [ [-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
 # If all the digits in the filter don't add up to 0 or 1, you
# should probably do a weight to get it to do so
# so, for example, if your weights are 1,1,1 1,2,1 1,1,1
# They add up to 10, so you would set a weight of .1 if you want to normalize them
weight  = 1
```

```python
i_transformed = np.copy(i).astype(np.int32) # Convert to signed integer type
size_x = i_transformed.shape[0]
size_y = i_transformed.shape[1]

for x in range(1,size_x-1):
  for y in range(1,size_y-1):
      output_pixel = 0.0
      output_pixel = output_pixel + (i_transformed[x - 1, y-1] * filter[0][0])
      output_pixel = output_pixel + (i_transformed[x, y-1] * filter[0][1])
      output_pixel = output_pixel + (i_transformed[x + 1, y-1] * filter[0][2])
      output_pixel = output_pixel + (i_transformed[x-1, y] * filter[1][0])
      output_pixel = output_pixel + (i_transformed[x, y] * filter[1][1])
      output_pixel = output_pixel + (i_transformed[x+1, y] * filter[1][2])
      output_pixel = output_pixel + (i_transformed[x-1, y+1] * filter[2][0])
      output_pixel = output_pixel + (i_transformed[x, y+1] * filter[2][1])
      output_pixel = output_pixel + (i_transformed[x+1, y+1] * filter[2][2])
      output_pixel = output_pixel * weight

      if(output_pixel<0):
        output_pixel=0
      if(output_pixel>255):
        output_pixel=255
      i_transformed[x, y] = output_pixel

# Convert back to uint8 and display the transformed image
i_transformed = i_transformed.astype(np.uint8)
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
plt.show()
```
